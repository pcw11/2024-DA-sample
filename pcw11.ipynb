### 데이터 사용

```python
import csv

inst_status_pwd_2023 = []

with open('../데이터/1996~2023연령별_인구변화_전체.csv', 'r', encoding='euc-kr') as f:
    data = csv.reader(f)

    next(data)

    for row in data:
        row_2023 = row[0:2] + row[18 : 27]
        inst_status_pwd_2023.append(row_2023)
        print(row_2023)
```

### 원그래프로 표현

```python
import matplotlib.pyplot as plt

data = []
name = ['어린이', '청소년', '청년', '중년', '노년']


plt.pie(data, colors = ['yellow', 'green', 'aqua', 'navy','puple'] )
plt.title('2023년 전국 인구 연령별 비율')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt

# 연령대별 비율 데이터 (주어진 비율)
data = [12.9, 11.6, 19.1, 17, 22.1]  # 각 연령대의 인구 비율

# 연령대 이름
name = ['어린이', '청소년', '청년', '중년', '노년']

# 파이차트 그리기
plt.pie(data, labels=name, colors=['yellow', 'green', 'aqua', 'navy', 'purple'], autopct='%1.1f%%', startangle=140)

# 그래프 제목
plt.title('2023년 전국 인구 연령별 비율')

# 범례 추가
plt.legend(name, loc="upper left")

# 원그래프 표시
plt.show()
```

```python
import matplotlib.pyplot as plt

# 데이터
sizes = [12.9, 100 - 12.9]  # 12.9%와 나머지 87.1%
labels = ['12.9%', '총 인구 비율 87.1%']

# 색깔 설정
colors = ['yellow', 'lightgrey']

# 원그래프 그리기
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=90)

# 타이틀 추가
plt.title('2023년 어린이 비율')

# 그래프 출력
plt.axis('equal')  # 원 모양으로 유지
plt.show()
```
